```python
import pandas as pd
df = pd.read_csv('total_data.csv')
print(df)
```

```python
df.drop_duplicates(subset='申请号', keep='first', inplace=True)
print(df)
```

```python
print(df.isnull().sum(axis=0))
```

```python
import cpca

df_addr = cpca.transform(df['地址'].tolist())

print(df_addr)
print(type(df_addr))
```

```python
df_province = df_addr['省'].value_counts().reset_index()
df_province.columns = ['省', 'counts']
df_province = df_province.drop([0])
print(df_province)
print(type(df_province))
```

```python
df_province['省'] = df_province['省'].str[0:2]
print(df_province)
df_province.to_csv('./data_of_visual/province_aply_count.csv', index=False, header=True)
```
