# 1. 申请人——排行榜

```python
import pandas as pd
df = pd.read_csv('total_data.csv')
print(df)
```

```python
df.drop_duplicates(subset='申请号', keep='first', inplace=True)
print(df)
```

```python
print(df.isnull().sum(axis=0))
```

```python
import pandas as pd
dff = pd.DataFrame([{'var1': 'a', 'var2': 1, 'var3':1},
                   {'var1': 'a,d,e', 'var2': 2, 'var3':1}])
                  
dfff = pd.concat([pd.Series(row['var2'], row['var1'].split(',')) for _, row in dff.iterrows()]).reset_index()
dfff.columns = ['var1', 'var2']
print(dfff)
```

```python
df1 = pd.concat([pd.Series(row['专利类型'], row['申请人'].split(';')) for _, row in df.iterrows()]).reset_index()
df1.columns = ['申请人', '专利类型']
print(df1)
```

```python
total_pat_rank = df1['申请人'].value_counts().reset_index()
total_pat_rank.columns = ['申请人', 'total_counts']
df_top10 = total_pat_rank.iloc[:10]
df_top10.sort_values('申请人', inplace=True)
print(type(df_top10))
print(df_top10)
```

```python
apply_top10 = total_pat_rank['申请人'][:10].tolist()
print(type(apply_top10))
print(apply_top10)
```

```python
gp = df1.groupby(by=['申请人','专利类型'])
df1_1 = gp.size()
cla_pat_rank = df1_1.reset_index(name='counts')
print(cla_pat_rank)
```

```python
cla_pat_rank.sort_values('counts',inplace=True, ascending=False)
print(type(cla_pat_rank))
print(cla_pat_rank)
```

```python
loc_top10 = cla_pat_rank.loc[cla_pat_rank['申请人'].isin(apply_top10)]
print(type(loc_top10))
print(loc_top10)
```

```python
df_out = loc_top10.set_index(['申请人',loc_top10.groupby(['申请人']).cumcount()+1]).unstack().sort_index(level=1, axis=1)
df_out.columns = df_out.columns.map('{0[0]}_{0[1]}'.format)
df_out = df_out.reset_index()
# df_out.sort_values('申请人', inplace=True)
print(type(df_out))
print(df_out.columns)
```

```python
df_out.sort_values('申请人', inplace=True)
print(df_out)
```

```python
result = pd.merge(df_out, df_top10)
result.sort_values('total_counts', inplace=True, ascending=False)
result.fillna(0, inplace=True)
for i in result.columns:
    if i[:6] == 'counts':
        result[i] = result[i].astype(int)
print(result)
result.to_csv('./data_of_visual/applicant_rank.csv', index=False, header=True)
```

# 3. 申请人——技术领域对比

```python
df3 = pd.concat([pd.Series(row['主分类号'], row['申请人'].split(';')) for _, row in df.iterrows()]).reset_index()
df3.columns = ['申请人', '主分类号']
print(df3)
```

```python
df3['分类'] = df3['主分类号'].str[0]
print(df3)
```

```python
df3.drop('主分类号', axis=1, inplace=True)
print(df3)
```

```python
df3 = df3.loc[df3['申请人'].isin(apply_top10)]
gp3_for_map = df3.groupby(['分类', '申请人'])
df3_for_map = gp3_for_map.size()
class_count_for_map = df3_for_map.reset_index(name='counts')
print(class_count_for_map)
```

```python
values = list(set(class_count_for_map['分类'].values))
print(values)
total_class = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
for i in total_class:
    if i not in values:
        print(i)
        new=pd.DataFrame({'分类': i,
                          '申请人': '0',
                          'counts':'0'},
                            index=[1]) 
        insert_empty_class = class_count_for_map.append(new, ignore_index=True) 
insert_empty_class.sort_values('分类', inplace=True)
order = ['分类', '申请人', 'counts']
insert_empty_class = insert_empty_class[order]
print(insert_empty_class)
```

```python
df_out = insert_empty_class.set_index(['分类',insert_empty_class.groupby(['分类']).cumcount()+1]).unstack().sort_index(level=1, axis=1)
df_out.columns = df_out.columns.map('{0[0]}_{0[1]}'.format)
df_out = df_out.reset_index()
# df_out.sort_values('申请人', inplace=True)
print(df_out)
```

```python
result = df_out
result.fillna(0, inplace=True)
for i in result.columns:
    if i[:6] == 'counts':
        result[i] = result[i].astype(int)
print(result)
```

```python
result.drop('分类', axis=1, inplace=True)
class_name = ['A(农业)', 'B(作业；运输)', 'C(化学；冶金)', 'D(纺织；造纸)', 
              'E(固定建筑物)', 'F(机械工程、照明、加热)', 'G(物理)', 'H(电学)']
result.insert(0, '分类', class_name) 
print(result)
result.to_csv('./data_of_visual/apply_tec_field_map.csv', index=False, header=True)
```
