```python
import pandas as pd
import numpy as np
with open('./outputfilename.csv') as f:
    df = pd.read_csv(f)
```

```python
df.dropna(subset=['tagname'], inplace=True)
df = df.sort_values(['tagname', 'time'])
df = df.astype(str)
print(df['qdyyb'].value_counts())
print(df)
```

```python
date_start =df["time"].min()
date_end = df["time"].max()
datetime_imput = [] 
date_start=pd.to_datetime(date_start)
date_end=pd.to_datetime(date_end)
i = date_start
j = date_end
while  i <= j:
    i = i + pd.Timedelta("1s")
    datetime_imput.append(i)
data_1 = pd.DataFrame(datetime_imput)
data_1["time1"]=data_1
```

```python
dataframe1=data_1.merge(df, how = 'left',left_on='time1',right_on='time')
print(dataframe1['zxtyy'].value_counts())
```
