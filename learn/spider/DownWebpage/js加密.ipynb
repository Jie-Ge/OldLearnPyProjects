- salt、sign是加密信息
- 1. 这个是计算salt的公式： i = "" + (new Date).getTime() + parseInt(10 * Math.random(), 10);
- 2. 计算sign：n.md5("fanyideskweb" + e + i + "n%A-rKaT5fb[Gy?;N5@Tj")
- 3. md5共需要4个参数，第一个和第4个都是固定值的字符串，第3个是salt，第二个参数是输入的要查找的单词

```python
# 版本02：获取salt、sign

def getSalt():
    '''
    salt的公式： i = "" + (new Date).getTime() + parseInt(10 * Math.random(), 10)
    将其翻译成python代码
    
    可以在浏览器的控制台中输入上面的函数，看结果，再写python代码
    
    '''
    import time, random
    
    salt = int(time.time()*1000) + random.randint(0,10)
    
    return salt

def getMD5(v):
    import hashlib
    
    md5 = hashlib.md5()
    
    md5.update(v.encode('utf-8'))
    sign = md5.hexdigest()
    
    return sign

def getSign(key, salt):
    
    sign = "fanyideskweb" + key + str(salt) + "n%A-rKaT5fb[Gy?;N5@Tj"
    
    sign = getMD5(sign)
    
    return sign

from urllib import request, parse

'''
破解有道翻译（爬有道翻译）
'''

def youdao(key):
    
    url = "http://fanyi.youdao.com/translate_o?smartresult=dict&smartresult=rule"
    
    salt = getSalt() # 确保是同一把‘盐’计算出的sign
    
    data = {
        "i": key,
        "from": "AUTO",
        "to": "AUTO",
        "smartresult": "dict",
        "client": "fanyideskweb",
        
        # ‘盐’：很长的随机串，将明文加上‘盐’串，再加密成密文，这样，就不容易破解了
        "salt": str(salt),
        "sign": getSign(key, salt),
        
        "ts": "1581053118819",
        "bv": "75a84f6fbcebd913f0a4e81b6ee54608",
        "doctype": "json",
        "version": "2.1",
        "keyfrom": "fanyi.web",
        "action": "FY_BY_REALTlME"
    }
    
    # 参数data需要是bytes格式
    data = parse.urlencode(data).encode()
    
    headers = {
        "Accept": "application/json, text/javascript, */*; q=0.01",
#         "Accept-Encoding": "gzip, deflate",
        "Accept-Language": "zh-CN,zh;q=0.9",
        "Connection": "keep-alive",
        "Content-Length": str(len(data)), # 改
        "Content-Type": "application/x-www-form-urlencoded; charset=UTF-8",
        "Cookie": "OUTFOX_SEARCH_USER_ID=1968066138@10.168.1.247; OUTFOX_SEARCH_USER_ID_NCOO=1126051056.209673; P_INFO=bysj02@163.com|1553081046|0|other|11&9|sic&1553077052&carddav#sic&510700#10#0#0|137316&0|mail163&blog|bysj02@163.com; _ntes_nnid=4e798d237af7e32f3b233793bbc58443,1576308921924; JSESSIONID=aaaMtgXXIHA3a-u09gEax; ___rl__test__cookies=1581053118814",
        "Host": "fanyi.youdao.com",
        "Origin": "http://fanyi.youdao.com",
        "Referer": "http://fanyi.youdao.com/",
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.117 Safari/537.36",
        "X-Requested-With": "XMLHttpRequest"
    }
    
    req = request.Request(url=url, data=data, headers=headers)
    
    rsp = request.urlopen(req)
    
    html = rsp.read().decode()
    
    print(html)
    
if __name__ == '__main__':
    youdao("boy")
```
