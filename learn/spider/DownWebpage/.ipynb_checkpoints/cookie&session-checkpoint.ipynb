- cookie & session
    - 由于http协议的无记忆性，人们为了弥补这个缺憾，所采用的一个补充协议
    - 例如：有了他们，当同一个浏览器再次访问一个需要登录的网站时，就不需要登录了，浏览器会通过他们验证用户的信息
    - cookie是发给用户（即http浏览器）的一段信息
    - session是保存在服务器上的对应的另一半信息，用来记录用户信息
    - Cookie通过在客户端记录信息确定用户身份，Session通过在服务器端记录信息确定用户身份
- cookie和session的区别：
    - 存放位置不同，cookie是在用户端，session是在服务器端（一般是在内存或者数据库中）且会过期
    - cookie是在用户端，所以不安全
    - 单个cookie保存数据不超过4k，很多浏览器限制一个站点最多保存20个

- 未使用cookie，【案例01】,是未登录状态
- 手动使用cookie（复制粘贴），【案例02】，登录状态
- 自动使用cookie
    - http模块包含一些关于cookie的模块，通过他们可以自动使用cookie
        - CookieJar
            - 管理存储cookie，向传出的http请求添加cookie
            - cookie存放在内存中，CookieJar实例回收后cookie将消失
        - FileCookieJar(filename, delayload=None, policy=None)
            - 使用文件管理cookie
            - filename是保存cookie的文件
            - MozillaCookieJar(filename, delayload=None, policy=None)
                - 创建与Mozilla浏览器cookie.txt兼容的FileCookieJar实例
            - LwpCookieJar(filename, delayload=None, policy=None)
                - 创建与libwww-perl标准兼容的Set-Cookie3格式的FileCookieJar实例
        - 关系：CookieJar --> FileCookieJar --> MozillaCookieJar --> LwpCookieJar
       
     - 一般CookieJar就够用了，【案例03】
         - 大致流程：
             1. 打开登录页面后自动通过用户名密码登录
             2. 自动提取反馈回来的cookie
             3. 利用提取的cookie登录隐私页面
    - cookie的属性
        - name
        - value
        - domain: 可以访问此cookie的域名
        - path：可以访问此cookie的页面路径
        - expires: 过期时间
        - size
        - http字段
    - cookie的保存：FileCookieJar 【案例04】
    - cookie的读取，案例【05】

```python
# 案例01： 未使用cookie

from urllib import request

url = "http://www.renren.com/973638553/profile"

req = request.Request(url)

rsp = request.urlopen(req)

html = rsp.read().decode('gbk', 'ignore')

with open("cookie1.html", "w") as f:
    f.write(html)
```

```python
# 案例03： 自动使用cookie

from urllib import request, parse
from http import cookiejar

# 创建cookiejar的实例
cookie = cookiejar.CookieJar()

# 生成cookie的管理器
cookie_handler = request.HTTPCookieProcessor(cookie)

# 创建http请求管理器
http_handler = request.HTTPHandler()

# 生成https管理器
https_handler = request.HTTPSHandler()

# 创建请求管理器
opener = request.build_opener(http_handler, https_handler, cookie_handler)

# 获取cookie
def login():
    '''
    负责初次登录
    需要输入用户名密码，用来获取登录cookie凭证
    '''
    
    # 此url需要从登录form的action属性中提取（查看网页源码中）
    url = "http://www.renren.com/PLogin.do"
    
    # 键值要与网页源码的form中的name一样
    data = {
        "email": "13119144223",
        "password":"123456"
    }
    
    # 吧数据进行编码
    data = parse.urlencode(data)
    
    # 创建一个请求对象
    req = request.Request(url, data=data.encode())
    
    # 使用opener发起请求
    rsp = opener.open(req)
    
def getHomePage():
    
    # 这是登录后隐私网页
    url = "http://www.renren.com/973638553/profile" 
    
    # 如果已经执行login函数，则opener会自动包含相应的cookie值
    rsp = opener.open(url)
    
    html = rsp.read().decode()

    with open("cookie3.html", "w") as f:
        f.write(html)
        
if __name__ == '__main__':
    login()
    
    # 看一下cookie
    print(cookie)
    for item in cookie:
        print('item: ', item)
        for i in dir(item): # dir()返回参数的属性、方法列表
#             print(i)
    
    getHomePage()
```

```python
# 案例04： 保存cookie

from urllib import request, parse
from http import cookiejar

# 创建filecookiejar的实例
filename = 'cookieSave.txt'
cookie = cookiejar.MozillaCookieJar(filename)

# 生成cookie的管理器
cookie_handler = request.HTTPCookieProcessor(cookie)

# 创建http请求管理器
http_handler = request.HTTPHandler()

# 生成https管理器
https_handler = request.HTTPSHandler()

# 创建请求管理器
opener = request.build_opener(http_handler, https_handler, cookie_handler)

# 获取cookie
def login():
    '''
    负责初次登录
    需要输入用户名密码，用来获取登录cookie凭证
    '''
    
    # 此url需要从登录form的action属性中提取（查看网页源码中）
    url = "http://www.renren.com/PLogin.do"
    
    # 键值要与网页源码的form中的name一样
    data = {
        "email": "13119144223",
        "password":"123456"
    }
    
    # 吧数据进行编码
    data = parse.urlencode(data)
    
    # 创建一个请求对象
    req = request.Request(url, data=data.encode())
    
    # 使用opener发起请求
    rsp = opener.open(req)
    
    # 保存cookie到文件
    # ignore_discard：即使cookie将要被丢弃也要保存
    # ignore_expires：如果该文件中的cookie即使已经过期，也保存
    cookie.save(ignore_discard=True, ignore_expires=True)
        
if __name__ == '__main__':
    login()
```

```python
# 案例05： cookie的读取

from urllib import request, parse
from http import cookiejar

# 创建cookiejar的实例
cookie = cookiejar.MozillaCookieJar()
cookie.load('cookieSave.txt', ignore_discard=True, ignore_expires=True)
    
def getHomePage():
    
    # 这是登录后隐私网页
    url = "http://www.renren.com/973638553/profile" 
    
    # 如果已经执行login函数，则opener会自动包含相应的cookie值
    rsp = opener.open(url)
    
    html = rsp.read().decode()

    with open("cookie5.html", "w") as f:
        f.write(html)
        
if __name__ == '__main__':
    getHomePage()
```
