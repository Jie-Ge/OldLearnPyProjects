- 正则常用方法
    - 参考开头的链接
    - match：从开始位置开始查找，一次匹配
    - search：从任何位置开始查找，一次匹配
    - findall：全部匹配，返回列表
    - finditer：全部匹配，返回迭代器
    - sub：替换
    - split：将字符串分割

```python
import re

#查找数字
#r表示字符串不转义
p = re.compile(r'\d+') # 生成一个正则表达式（ Pattern ）对象

m = p.match("one12two34", 3, 10) # 匹配

print(m)

print(m.group(0))
print(m[0])

print(m.start(0))
print(m.end(0))
```

```python
pattern = re.compile(r'([a-z]+) ([a-z]+)', re.I)   # re.I 表示忽略大小写
m = pattern.match('Hello World Wide Web')
print(m)
```

```python
# 或 print(m.group())
print(m.group(0)) # 返回匹配到的整个字符串
print(m.start(0))
print(m.end(0))
```

```python
print(m.group(1)) # 返回匹配到的第一个组
print(m.start(1))
print(m.end(1))
```

```python
print(m.groups())

print(type(m.group(0)))
print(type(m.groups()))
```

# 匹配中文
- 大部分中文内容的表示范围是【u4e00-u9fa5】，不包括全角标点

```python
import re

# u后面字符串以 Unicode 格式 进行编码，一般用在中文字符串前面，
# 防止因为源码储存格式问题，导致再次使用时出现乱码。
t = u'世界 你好，hello world'

p = re.compile(r'[\u4e00-\u9fa5]+')

r = p.findall(t)

print(r)
```
