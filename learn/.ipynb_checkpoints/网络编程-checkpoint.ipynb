# socket编程
- Socket又称"套接字"，应用程序通常通过"套接字"向网络发出请求或者应答网络请求，使主机间或者一台计算机上的进程间可以通讯。

- 通常我们用一个Socket表示“打开了一个网络链接”，而打开一个Socket需要知道目标计算机的IP地址和端口号，再指定协议类型即可。

- 通过 ‘ip+端口’ 定位对方

- 主动发起连接的叫客户端，被动响应连接的叫服务器

- 服务器城程序要求永久运行，一般用死循环处理

```python
# UDP编程案例

import socket

# 服务器要先启动，等待接收消息
def serverFunc():
    # 1. 建立socket
    
    # socket.AF_INET: 使用ipv4协议
    # socket.SOCK_DGRAM：使用UDP协议
    s = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
    
    # 2. 绑定 port 和 ip
    
    # 127.0.0.1：本机  ( 也可：host = socket.gethostname() # 获取本地主机名 )
    # 9999：随机的端口
    # 地址是一个元组类型: (ip, port)
    s.bind(('127.0.0.1', 9999))
    
    # 3. 接收对方信息（ 死等 ）
    
    # recvfrom接收的返回值是一个元组，前一项表示数据，后一项表示地址
    # 参数500表示缓冲区大小（指定要接收的最大数据量）
    data, addr = s.recvfrom(500)
    
    # 发送过来的数据是bytes格式，必须要通过解码才能得到str格式的内容
    text = data.decode() # 默认是utf-8
    print(type(text))
    print(text)
    
    # 4. 给对方发送反馈
    rsp = "I am not hunger"
    
    # 发送的数据需要编码成bytes格式 （也可直接用 rsp = b'I am not hunger'）
    data = rsp.encode() # 默认的是utf-8
    s.sendto(data, addr)
    
if __name__ == '__main__':
    print('Starting server .......')
    while True:
        serverFunc()
    print('Ending server......')
```
