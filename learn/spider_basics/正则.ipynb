```python
import re

#
#r
p = re.compile(r'\d+') #  Pattern 

m = p.match("one12two34", 3, 10) # 

print(m)

print(m.group(0))
print(m[0])

print(m.start(0))
print(m.end(0))
```

```python
pattern = re.compile(r'([a-z]+) ([a-z]+)', re.I)   # re.I 
m = pattern.match('Hello World Wide Web')
print(m)
```

```python
#  print(m.group())
print(m.group(0)) # 
print(m.start(0))
print(m.end(0))
```

```python
print(m.group(1)) # 
print(m.start(1))
print(m.end(1))
```

```python
print(m.groups())

print(type(m.group(0)))
print(type(m.groups()))
```

```python
import re

# u Unicode  
# 
t = u' hello world'

p = re.compile(r'[\u4e00-\u9fa5]+')

r = p.findall(t)

print(r)
```
