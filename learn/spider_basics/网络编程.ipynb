```python
# UDP

import socket

# 
def serverFunc():
    # 1. socket
    
    # socket.AF_INET: ipv4
    # socket.SOCK_DGRAMUDP
    s = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
    
    # 2.  port  ip
    
    # 127.0.0.1  ( host = socket.gethostname() #  )
    # 9999
    # : (ip, port)
    s.bind(('127.0.0.1', 9999))
    
    # 3.   
    
    # recvfrom
    # 500
    data, addr = s.recvfrom(500)
    
    # bytesstr
    text = data.decode() # utf-8
    print(type(text))
    print(text)
    
    # 4. 
    rsp = "I am not hunger"
    
    # bytes  rsp = b'I am not hunger'
    data = rsp.encode() # utf-8
    s.sendto(data, addr)
    
if __name__ == '__main__':
    print('Starting server .......')
    while True:
        serverFunc()
    print('Ending server......')
```
